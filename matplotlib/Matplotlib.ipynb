```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d=np.linspace(0,np.pi*4,100)
sin_y=np.sin(d)
cos_y=np.cos(d)
plt.plot(d,sin_y,color='#000',label='sin' )
plt.plot(d,cos_y,label='cos')
plt.legend(loc="upper right")

plt.title('sine wave')
plt.show()
```

# 4.   scatter plot

```python
n=100
x=np.random.rand(n)
y=np.random.rand(n)
plt.style.use('seaborn-v0_8-darkgrid')
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)

plt.scatter(x,y)
plt.show()
```

```python
n=1000
x=np.random.randn(n)
y=np.random.randn(n)

x2=np.random.randn(n)+5
y2=np.random.randn(n)+1

plt.scatter(x,y,c='red',alpha=.6,s=10,label='g1')
plt.scatter(x2,y2,c='blue',alpha=.6,s=10,label='g2')
plt.legend()
plt.show()
```

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=rng.rand(100)*1000

plt.scatter(x,y,c=colors,s=sizes,alpha=.7,cmap='viridis',edgecolors='green')
plt.title('random scatter plot')
plt.colorbar()
plt.show()
```

# pie charts

```python
d=['a','b','c','d','e','f','g']
data=[401,780,450,200,45,780,45]
plt.style.use('fast')
plt.pie(data,labels=d)
plt.show()
```

```python

plt.pie(data,labels=d,explode=[.1,.2,.1,.2,.35,.4,.3])
plt.show()
```

```python

plt.pie(data,labels=d,explode=[.1,.2,.1,.2,.35,.4,.3],shadow=True)
plt.show()
```

## adding parcentage

```python

plt.pie(data,labels=d,explode=[.1,.2,.1,.2,.35,.4,.3],shadow=True,autopct='%.1F%%')
plt.show()
```

```python
plt.pie(data,labels=d,explode=[.1,.2,.1,.2,.35,.4,.3],shadow=True,autopct='%.1F%%',
       wedgeprops={'edgecolor':'black','linewidth':2})


plt.show()


```

```python
plt.pie(data,labels=d,explode=[.1,.2,.1,.2,.35,.4,.3],shadow=True,autopct='%.1F%%',
       wedgeprops={'edgecolor':'black','linewidth':2,
                'antialiased':False})


plt.show()
```
